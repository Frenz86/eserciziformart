```python
import requests
from PIL import Image

url = "http://127.0.0.1:8000/download/dog.jpg"
response = requests.get(url)

if response.status_code == 200:
    with open("downloaded_dog.jpg", "wb") as f:
        f.write(response.content)
    print("Immagine scaricata con successo!")
    img = Image.open("downloaded_dog.jpg")
    display(img)
else:
    print("Errore durante il download dell'immagine:", response.status_code)
```
